```python
TDPTProbability = (gamma**2*((omega-omega0)**2)/4)*(np.sin(t*(omega-omega0)/2))**2
TDPTProbabilityN = preprocessing.normalize([TDPTProbability])
ExactProbability = (gamma**2)*((gamma**2+(omega-omega0)**2)/4)*(np.sin(t*np.sqrt((gamma**2+(omega-omega0)/2))))**2
#plt.plot(t,TDPTProbability)
plt.plot(t, ExactProbability)
```
